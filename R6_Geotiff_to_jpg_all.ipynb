```python
from osgeo import gdal
import os
import glob
```

```python
#フォルダ名
folder_path = '/content/drive/MyDrive/NorthKorea/'


# TIFFファイルのリストを取得
#上は、単独のGeotiff画像をjpg化する際に使用し、下はモザイク画像をjpg化する際に使う
tif_files = glob.glob(os.path.join(folder_path, '*_R.tif'))
#tif_files = glob.glob(os.path.join(folder_path, 'mosaic_*.tif'))

for input_geotiff in tif_files:
    # TIFFファイル名から拡張子を変更して、出力ファイル名を生成
    output_jpg = os.path.splitext(os.path.basename(input_geotiff))[0] + '.jpg'
    output_jpeg = os.path.join(folder_path, output_jpg)
    print(output_jpg)

    ds = gdal.Open(input_geotiff)

    options = gdal.TranslateOptions(
        format='JPEG',
        creationOptions=['QUALITY=70']
    )

    gdal.Translate(output_jpeg, ds, options=options)

    # ファイルを閉じる
    ds = None

print('変換が完了しました。')


```
